```python
records = []
for i in range(0,768):
    records.append([str(df.values[i,j]) for j in range(0,5)])
```

```python
for sets in records:
    print(sets)
```

```python
association_rules = apriori(records, min_support = 0.005, min_confidence = 0.07, min_lift = 3, min_length = 2)
```

```python
association_results = list(association_rules)
```
